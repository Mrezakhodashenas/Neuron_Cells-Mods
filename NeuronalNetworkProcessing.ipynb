```python
# @article{cutts2014detecting,
#   title={Detecting pairwise correlations in spike trains: an objective comparison of methods and application to the study of retinal waves},
#   author={Cutts, Catherine S and Eglen, Stephen J},
#   journal={Journal of Neuroscience},
#   volume={34},
#   number={43},
#   pages={14288--14303},
#   year={2014},
#   publisher={Soc Neuroscience}
# }

# @article{kreuz2013monitoring,
#   title={Monitoring spike train synchrony},
#   author={Kreuz, Thomas and Chicharro, Daniel and Houghton, Conor and Andrzejak, Ralph G and Mormann, Florian},
#   journal={Journal of neurophysiology},
#   volume={109},
#   number={5},
#   pages={1457--1472},
#   year={2013},
#   publisher={American Physiological Society Bethesda, MD}
# }

import numpy as np

def compute_STTC(spk_times_1, spk_times_2, dt, T):
    """
    Calculate the Spike Time Tiling Coefficient (STTC) between two spike trains.

    spk_times_1: list of spike times for the first neuron
    spk_times_2: list of spike times for the second neuron
    dt: time window (in ms) for considering two spikes as coincident
    T: total duration of the recording (in ms)
    """
    # Calculate the proportion of spikes in spk_times_1 that are within dt of a spike in spk_times_2
    N1 = len(spk_times_1)
    N2 = len(spk_times_2)

    if N1 == 0 or N2 == 0:
        return np.nan

    # Calculate P1: Proportion of spikes in spk_times_1 with a nearby spike in spk_times_2
    P1 = np.sum([np.any(np.abs(spk_times_2 - t1) <= dt) for t1 in spk_times_1]) / N1

    # Calculate P2: Proportion of spikes in spk_times_2 with a nearby spike in spk_times_1
    P2 = np.sum([np.any(np.abs(spk_times_1 - t2) <= dt) for t2 in spk_times_2]) / N2

    # Calculate T_A: Proportion of the recording duration within dt of any spike in spk_times_1 or spk_times_2
    TA1 = 2 * N1 * dt / T
    TA2 = 2 * N2 * dt / T

    # Compute the STTC value
    STTC = 0.5 * ((P1 - TA1) / (1 - P1 * TA1) + (P2 - TA2) / (1 - P2 * TA2))

    return STTC

```

```python

# Function to calculate population synchrony
def calculate_population_synchrony(spkt_filtered, spkid_filtered, dt, T):
    unique_ids = np.unique(spkid_filtered)
    num_neurons = len(unique_ids)
    sttc_values = []

    for i in range(num_neurons):
        spk_times_1 = np.array([spkt_filtered[j] for j in range(len(spkid_filtered)) if spkid_filtered[j] == unique_ids[i]])
        for j in range(i + 1, num_neurons):
            spk_times_2 = np.array([spkt_filtered[k] for k in range(len(spkid_filtered)) if spkid_filtered[k] == unique_ids[j]])
            sttc = compute_STTC(spk_times_1, spk_times_2, dt, T)
            if not np.isnan(sttc):
                sttc_values.append(sttc)

    return np.mean(sttc_values) if sttc_values else np.nan

# Extract all spike times and spike IDs
spkt = data['simData']['spkt']
spkid = data['simData']['spkid']

# Define the neuron ID ranges for each population
populations = {
    'EC_Pyr': range(0, 15),
    'GC_Pyr': range(15, 30),
    'CA3_Pyr': range(30, 45),
    'CA1_Pyr': range(45, 60),
    'CA1_int': range(60, 69)
}

# Time window and recording duration parameters for STTC
dt = 10  # Time window for STTC calculation in ms
T = max(spkt)  # Total duration of the recording

# Function to filter spikes by population
def filter_spikes(pop_name):
    pop_range = populations[pop_name]
    spkt_filtered = [spkt[i] for i in range(len(spkid)) if spkid[i] in pop_range]
    spkid_filtered = [spkid[i] for i in range(len(spkid)) if spkid[i] in pop_range]
    return spkt_filtered, spkid_filtered

# Plot raster plot for each population and calculate synchrony
for pop_name in populations.keys():
    spkt_filtered, spkid_filtered = filter_spikes(pop_name)

    # Calculate synchrony for the population
    sync_measure = calculate_population_synchrony(spkt_filtered, spkid_filtered, dt, T)
    print(f'Synchrony measure for {pop_name}: {sync_measure}')
```

### Cross-population Synchrony

```python
# @article{battaglia2012dynamic,
#   title={Dynamic effective connectivity of inter-areal brain circuits},
#   author={Battaglia, Demian and Witt, Annette and Wolf, Fred and Geisel, Theo},
#   journal={PLoS computational biology},
#   volume={8},
#   number={3},
#   pages={e1002438},
#   year={2012},
#   publisher={Public Library of Science San Francisco, USA}
# }

# @article{battaglia2012dynamic,
#   title={Dynamic effective connectivity of inter-areal brain circuits},
#   author={Battaglia, Demian and Witt, Annette and Wolf, Fred and Geisel, Theo},
#   journal={PLoS computational biology},
#   volume={8},
#   number={3},
#   pages={e1002438},
#   year={2012},
#   publisher={Public Library of Science San Francisco, USA}
# }

# Function to filter spikes by population
def filter_spikes(pop_name, populations, spkt, spkid):
    pop_range = populations[pop_name]
    spkt_filtered = [spkt[i] for i in range(len(spkid)) if spkid[i] in pop_range]
    spkid_filtered = [spkid[i] for i in range(len(spkid)) if spkid[i] in pop_range]
    return spkt_filtered, spkid_filtered

# Parameters for STTC calculation
dt = 10  # Time window for STTC calculation in ms
T = max(spkt)  # Total duration of the recording

# Extract spikes for CA1_Pyr and CA3_Pyr
spkt_CA1, spkid_CA1 = filter_spikes('CA3_Pyr', populations, spkt, spkid)
spkt_CA3, spkid_CA3 = filter_spikes('CA3_Pyr', populations, spkt, spkid)

# Calculate cross-population synchrony between CA1_Pyr and CA3_Pyr
sttc_values = []

# Iterate through each neuron in CA1_Pyr and CA3_Pyr to compute STTC
for id_CA1 in np.unique(spkid_CA1):
    spk_times_CA1 = np.array([spkt_CA1[i] for i in range(len(spkid_CA1)) if spkid_CA1[i] == id_CA1])
    for id_CA3 in np.unique(spkid_CA3):
        spk_times_CA3 = np.array([spkt_CA3[i] for i in range(len(spkid_CA3)) if spkid_CA3[i] == id_CA3])
        sttc = compute_STTC(spk_times_CA1, spk_times_CA3, dt, T)
        if not np.isnan(sttc):
            sttc_values.append(sttc)

# Calculate the average STTC for cross-population synchrony
cross_pop_sync = np.mean(sttc_values) if sttc_values else np.nan
print(f'Cross-population synchrony between CA3_Pyr and CA3_Pyr: {cross_pop_sync}')
```
